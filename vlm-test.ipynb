This code requires having ollama setup on the desktop + llava:7b installed. The results from the code can be seen below and they are very simple so it is not worth running this code directly if you have setup this notebook locally.

```python
import base64
import ollama

from langchain_ollama.llms import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate

model = OllamaLLM(
    model="llava:7b",
    temperature=0
)

prompt = f"""
This is a small screenshot of a GUI icon from a design program like Figma, Illustrator or Inkscape.
Please output a viable label for this icon that reflects it's purpose. 

ONLY OUTPUT THE LABEL AS A STRING
Do not include anything else.
"""

def encode_image(path):
    with open(path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf-8")
```

```python
pen = encode_image("icon-datasets/labelled-toolbar-dataset/pen.png")
actions = encode_image("icon-datasets/labelled-toolbar-dataset/actions.png")
comment = encode_image("icon-datasets/labelled-toolbar-dataset/comment.png")
design = encode_image("icon-datasets/labelled-toolbar-dataset/design.png")
dev = encode_image("icon-datasets/labelled-toolbar-dataset/dev.png")
draw = encode_image("icon-datasets/labelled-toolbar-dataset/draw.png")
frame = encode_image("icon-datasets/labelled-toolbar-dataset/frame.png")
move = encode_image("icon-datasets/labelled-toolbar-dataset/move.png")
text = encode_image("icon-datasets/labelled-toolbar-dataset/text.png")

icons = []

icons = {
    "pen": pen, 
    "actions": actions, 
    "comment": comment, 
    "design": design, 
    "dev": dev, 
    "draw": draw, 
    "frame": frame, 
    "move": move, 
    "text": text
}

for icon, image in icons.items():
    
    response = ollama.chat(
    model="llava:7b",
    messages=[
        {
            "role": "user",
            "content": f"{prompt}",
            "images": [image]
        }
    ]
    )
    
    print(f'true label: {icon}; predicted label: {response["message"]["content"]}')
```

As can be seen from results - although this is more generalisable than SIFT it is significantly less accurate.
